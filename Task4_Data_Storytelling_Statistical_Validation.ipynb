```python
male_sales = df[df["Gender"] == "Male"]["Total_Sales"]
female_sales = df[df["Gender"] == "Female"]["Total_Sales"]
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(male_sales, female_sales)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

# Hypothesis Testing

## Objective
To determine whether there is a significant difference in Total Sales between Male and Female customers.

## Null Hypothesis (H₀)
There is no significant difference in average Total Sales between Male and Female customers.

## Alternative Hypothesis (H₁)
There is a significant difference in average Total Sales between Male and Female customers.

## Results
T-Statistic: 0.682
P-Value: 0.495

## Conclusion
Since the p-value (0.495) is greater than 0.05, we fail to reject the null hypothesis.

Therefore, there is no statistically significant difference in Total Sales between Male and Female customers.
